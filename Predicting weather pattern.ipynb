<div class="alert alert-block alert-warning">
This graph shows seasonality of daily maximum temperature in recent 3 years. Which makes sense since we always have peak value of max temperature around July of each year, and temperature drops down to the bottom around end of December in each year.

```python
min_temp = data4.min_temp.astype(int)
mean(min_temp)
max(min_temp)
min(min_temp)
%pylab inline
pylab.plot(data4.index,min_temp,'-b', label='min temperature')
pylab.legend(loc='upper right')
pylab.ylabel('Temperature')
```

<div class="alert alert-block alert-warning">
This graph is showing the similar seasonality as max temperature plot since in reality min temperature usually drops or rises along with max temperature drops or rises.
    </div>

<div class="alert alert-block alert-warning">
In precip graph, we can't really see an obvious seasonality for precipitation.
    </div>

<div class="alert alert-block alert-warning">
Again, although we can see that snow season usually starts around October and ends around late April, we can't find an apparent seasonality for snow fall.
